```python
import wooldridge
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
corn = wooldridge.data('corn')
```

```python
corn.head()
```

```python
corn_cor = corn[['cornhec', 'soyhec', 'cornpix', 'soypix']]
```

```python
sns.heatmap(corn_cor.corr(),annot = True)
```

```python
sns.pairplot(corn_cor)
```

```python
 
from  sklearn.cluster  import  KMeans
corn_Cluster = KMeans(n_clusters=3)
```

```python
corn_Cluster.fit(corn_cor)
```

```python
labels = corn_Cluster.labels_
labels = labels.reshape(37,1)
```

```python
corn['label'] = labels
```

```python
sns.pairplot(corn, vars=corn.columns[:-1], hue="label", palette='summer')
```
